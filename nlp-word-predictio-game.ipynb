```python
# NLP-Based Word Prediction Game (Hangman Style)
# Author: Barath P

import random

# Dataset: words with hints
dataset = {
    "python": "A popular programming language",
    "machine": "Used in machine learning",
    "language": "Used for communication",
    "science": "Field involving data and experiments",
    "algorithm": "Step-by-step problem solving method"
}

# Select random word and hint
word, hint = random.choice(list(dataset.items()))
word = word.lower()

guessed_letters = []
attempts = 6
display_word = ["_"] * len(word)

print("ğŸ¯ Welcome to the Word Prediction Game!")
print("Hint:", hint)
print("Word:", " ".join(display_word))

# Game loop
while attempts > 0 and "_" in display_word:
    guess = input("\nGuess a letter: ").lower()

    if not guess.isalpha() or len(guess) != 1:
        print("Please enter a single valid letter.")
        continue

    if guess in guessed_letters:
        print("You already guessed that letter.")
        continue

    guessed_letters.append(guess)

    if guess in word:
        print("âœ… Correct guess!")
        for i in range(len(word)):
            if word[i] == guess:
                display_word[i] = guess
    else:
        attempts -= 1
        print("âŒ Wrong guess! Attempts left:", attempts)

    print("Word:", " ".join(display_word))

# Game result
if "_" not in display_word:
    print("\nğŸ‰ Congratulations! You guessed the word:", word)
else:
    print("\nğŸ’€ Game Over! The word was:", word)
```
